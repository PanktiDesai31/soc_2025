```python
import pygame as pg
pg.init()
import numpy as np

class Healing(): #healing class, contains agent. goes to location. if cell is not healthy (green), it is healed. 
    def __init__(self, screen, cell_size, x, y): #constructor for the class, contains the vairables necessary in the class
        self.screen = screen
        self.cell_size = cell_size
        self.row = int(y/cell_size) #row index
        self.col = int(x/cell_size)
    
    def agent(self, b) : #healing agent for the grid
        if b[self.row][self.col] != 1 : #first it shows blue = the cell is healing then green = healed
            pg.draw.rect(self.screen, (0,0,255), pg.Rect(self.col * self.cell_size, self.row * self.cell_size, self.cell_size, self.cell_size))
            pg.display.update()
            pg.time.delay(2000)
            pg.draw.rect(self.screen, (0,255,0), pg.Rect(self.col * self.cell_size, self.row * self.cell_size, self.cell_size, self.cell_size))
            pg.display.update()

def drawgrid(screen, rows, cell_size, b): # function to draw the tissue sample. 
        for row in range(rows) :
            for column in range(rows):
                health = b[row][column] #checking the health at the grid square
                red = int(255 * health) #if health = 0, red = 0. healthy
                green = int(255 * (1 - health))
                color = (red, green, 0) # if healthy = (0,255,0) = green
                pg.draw.rect(screen, color, pg.Rect(column * cell_size, row * cell_size, cell_size, cell_size))
                pg.draw.rect(screen, (0,0,0), pg.Rect(column * cell_size, row * cell_size, cell_size, cell_size),1)
        pg.display.update()

def updatedgrid(screen, cell_size, change, x, y, b): # after we have altered the grid. clicking the cell makes it healthy or unhealthy. 
    col_index = int(x / cell_size); row_index = int(y / cell_size) #row and column wise position 
    color = screen.get_at((x, y))[:3]
    #gives us current health of cell.the first position(the red part) tells us if cell is healthy or not. 
    #can do the same with green, but closer to 1=healthy in that case
    health = color[0] / 255.0 
    #v r considering helath that's not just 0/1, in case we want to change a cell that's already been modified
    if health > 0.5: #unhealthy, becasue it's close to red
        new_health = max(0, health - change / 1000) #if its possible to get unhealthier, it does. 
    else:
        new_health = min(1, health + change / 1000)

    b[row_index][col_index] = new_health#changing health of cell in a mathematical way too. easier for calcn

    # draw the updated cell, with new health. 
    red = int(255 * new_health)
    green = int(255 * (1 - new_health))
    color = (red, green, 0)

    pg.draw.rect(screen, color, pg.Rect(col_index * cell_size, row_index * cell_size, cell_size, cell_size))
    pg.draw.rect(screen, (0, 0, 0), pg.Rect(col_index * cell_size, row_index * cell_size, cell_size, cell_size), 1)
    pg.display.update()
def main():
    a = np.random.randint(0,2, size = 100)
    b = a.reshape(10,10) 
    rows=10 #considering a square grid, columns = rows
    window_size = 500
    cell_size = window_size/rows 
    screen = pg.display.set_mode((window_size, window_size)) #represents my tissue grid.
    title = pg.display.set_caption("tissue sample")

    drawgrid(screen, rows, cell_size, b)
   
    change = 100 #how much does the health of the cell change. healthy becomes unhealthy and vice versa
    running = True
    while running:
        for event in pg.event.get():
            if event.type == pg.QUIT:
                running = False
            elif event.type == pg.MOUSEBUTTONDOWN:
                (x,y) = pg.mouse.get_pos() 
                print(x, y)
                updatedgrid(screen, cell_size, change, x, y, b) #first we see the damaged cell      
                pg.display.update()
                pg.time.delay(1000)
                healer = Healing(screen, cell_size, x, y) 
                healer.agent(b) #then we see it healing 

    pg.quit()
if __name__ == "__main__":
    main()
```
